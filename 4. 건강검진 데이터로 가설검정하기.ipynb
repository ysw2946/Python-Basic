## 상관분석
###  상관분석 구하기

```python
columns = ['연령대 코드(5세단위)', '신장(5Cm단위)','체중(5Kg 단위)', '허리둘레', '시력(좌)', '시력(우)', '청력(좌)', '청력(우)', '수축기 혈압',
       '이완기 혈압', '식전혈당(공복혈당)', '총 콜레스테롤', '트리글리세라이드', 'HDL 콜레스테롤', 'LDL 콜레스테롤',
       '혈색소', '요단백', '혈청크레아티닌', '(혈청지오티)AST', '(혈청지오티)ALT', '감마 지티피', '흡연상태',
       '음주여부']
columns
```

```python
df_small = df_sample[columns]
df_corr = df_small.corr()
df_corr
```

```python
df_corr.loc[df_corr["신장(5Cm단위)"] > 0.1, "신장(5Cm단위)"]
```

```python
df_corr["흡연상태"].sort_values()
```

```python
df_corr.loc[df_corr["흡연상태"] > 0.25, "흡연상태"]
```

```python
df_corr["혈색소"].sort_values(ascending=False).head(7)
```

### heatmap
